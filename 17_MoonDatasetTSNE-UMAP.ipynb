```python
import pandas as pd
import numpy as np
import umap
from sklearn.manifold import TSNE
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
x, y = make_moons(n_samples=500, noise=0.01)
x
```

```python
y
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()
```

```python
tsne = TSNE(n_components=1, random_state=123)
tsne_result = tsne.fit_transform(x)
tsne_result
```

```python
plt.scatter(tsne_result[:, 0], np.ones(len(tsne_result)), c=y)
plt.show()
```

```python
umap = umap.UMAP(n_components=1)
umap.fit(x)
umap_result = umap.transform(x)
umap_result
```

```python
plt.scatter(umap_result[:, 0], np.ones(len(umap_result)), c=y)
plt.show()
```
