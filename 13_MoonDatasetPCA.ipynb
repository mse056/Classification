```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
x, y = make_moons(n_samples=500, noise=0.01)
x
```

```python
y
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()
```

```python
pca = PCA(n_components=1)
pca.fit(x)
pca_result = pca.transform(x)
pca_result
```

```python
plt.scatter(pca_result[:, 0], np.ones(len(pca_result)), c=y)
plt.show()
```

```python
kpca = KernelPCA(n_components=1, kernel='poly')
kpca.fit(x)
kpca_result = kpca.transform(x)
kpca_result
```

```python
plt.scatter(kpca_result[:, 0], np.ones(len(kpca_result)), c=y)
plt.show()
```
